```
import os

# All paths of the dataset
path_normal    = "/content/drive/My Drive/Colab Notebooks/Breast cancer/dataset/train/normal/"
path_malignant = "/content/drive/My Drive/Colab Notebooks/Breast cancer/dataset/train/malignant/"
path_benign    = "/content/drive/My Drive/Colab Notebooks/Breast cancer/dataset/train/benign/"

list_original_path = [path_benign, path_malignant, path_normal]

# The number of images in the raw dataset
num_of_raw_images_benign    = len(os.listdir(list_original_path[0]))
num_of_raw_images_malignant = len(os.listdir(list_original_path[1]))
num_of_raw_images_normal    = len(os.listdir(list_original_path[2]))


num_of_raw_images = [num_of_raw_images_benign, num_of_raw_images_malignant, num_of_raw_images_normal]

print(num_of_raw_images)
# Calculate the images that will be used for augmentation in each type
def calculate_num_images_used():
  min_number = min(num_of_raw_images)
  total_images_created = min_number * 13
  index_min_number = num_of_raw_images.index(min_number)
  num_of_images_used = []
  for i in range(len(num_of_raw_images)):
    if i == index_min_number:
      num_of_images_used.append(num_of_raw_images[i])
    else:
      num_of_images_used.append((total_images_created - num_of_raw_images[i]) // 12)
  return num_of_images_used

print(calculate_num_images_used())
```
