```python
from google.colab import drive
import os
import random
import shutil

drive.mount('/content/drive')

raw_dataset_path = '/content/drive/My Drive/Colab Notebooks/Breast cancer/dataset_ori'
dataset_path = '/content/drive/My Drive/Colab Notebooks/Breast cancer/dataset'

img_type = os.listdir(raw_dataset_path)
img_each_type = []

for i in range(len(img_type)):
  img_each_type.append(len(os.listdir(os.path.join(raw_dataset_path ,img_type[i]))))

split_name_dir = ["train", "test", "val"]

for i in split_name_dir:
  os.makedirs(os.path.join(dataset_path, i))
  for j in img_type:
    os.makedirs(os.path.join(dataset_path, i, j))

list_img_benign    = os.listdir(os.path.join(raw_dataset_path, img_type[0]))
list_img_normal    = os.listdir(os.path.join(raw_dataset_path, img_type[1]))
list_img_malignant = os.listdir(os.path.join(raw_dataset_path, img_type[2]))

list_img = [list_img_benign, list_img_normal, list_img_malignant]

split_ratio = [0.2, 0.4, 0.4]

random.shuffle(list_img_benign)
random.shuffle(list_img_malignant)
random.shuffle(list_img_normal)

def image_split(list_img, ratio):
  num_train_split = int(len(list_img) * ratio[0])
  num_test_split  = int(len(list_img) * (ratio[0] + ratio[1]))
  train_img_list  = list_img[:num_train_split]
  test_img_list   = list_img[num_train_split:num_test_split]
  val_img_list    = list_img[num_test_split:]
  return train_img_list, test_img_list, val_img_list


for i in range(len(img_type)):
  train_list, test_list, val_list = image_split(list_img[i], split_ratio)
  for img in train_list:
    raw_img_path = os.path.join(raw_dataset_path, img_type[i], img)
    dest_path = os.path.join(dataset_path, split_name_dir[0], img_type[i], img)
    shutil.copyfile(raw_img_path, dest_path)

  for img in test_list:
    raw_img_path = os.path.join(raw_dataset_path, img_type[i], img)
    dest_path = os.path.join(dataset_path, split_name_dir[1], img_type[i], img)
    shutil.copyfile(raw_img_path, dest_path)

  for img in val_list:
    raw_img_path = os.path.join(raw_dataset_path, img_type[i], img)
    dest_path = os.path.join(dataset_path, split_name_dir[2], img_type[i], img)
    shutil.copyfile(raw_img_path, dest_path)

  print("Finished copying " + str(img_each_type[i]) + " of class " + img_type[i] + ".")
  print(str(len(train_list)) + " images used for training.")
  print(str(len(test_list)) + " images used for testing.")
  print(str(len(val_list)) + " images used for validation.\n")
```
